### 답안

```python
def solution(s):
  str_list = list(s)
  dic = {}
  answer = []

  for i, v in enumerate(list(s)):
    if (dic.get(v) == None):
      dic[v] = i
      answer.append(-1)
    else:
      answer.append(i - dic[v])
      dic[v] = i
  return answer


print(solution('banana'))
```

```python
def solution(s):
  str_list = list(s)

  dic = {}
  answer = []

  for i, v in enumerate(list(s)):

    # dictionary 키를 알파벳
    # dic.get(값) -> 없는 경우 None
    # dic[값] -> 없는 경우 오류
    if (dic.get(v) == None):
      dic[v] = i
      # {v: i}로 저장
      # 알파벳: 인덱스 (키는 중복 불가.)
      answer.append(-1)
      # 같은게 없기 때문에 -1 삽입
    else:
      answer.append(i - dic[v])
      # 현재위치 - 원래있던 위치 -> 몇번째 전을
      dic[v] = i
      # 새롭게 인덱스 위치 저장
  return answer

solution('banana')
```
