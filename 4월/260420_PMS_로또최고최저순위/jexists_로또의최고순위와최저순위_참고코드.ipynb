### 내 답안

```python
def solution(lottos, win_nums):
    win_count = 0
    unknown_count = 0

    winning_rank = {6:1, 5:2, 4:3, 3:4, 2: 5}
    for lotto in lottos:
        if (lotto == 0):
            unknown_count += 1
        for win in win_nums:
                    # print(win, lotto)
            if (win == lotto):
                win_count += 1


    answer = [winning_rank.get(win_count+unknown_count, 6), winning_rank.get(win_count, 6)]
    return answer


lottos = [44, 1, 0, 0, 31, 25]
win_nums = [31, 10, 45, 1, 6, 19]
solution(lottos, win_nums)
```

###### 잘된 점:

- 로직 자체는 정확함 — unknown_count로 0 처리, win_count로 일치 수 계산
- winning_rank.get(..., 6) 으로 기본값 6 처리도 깔끔

###### 버그/비효율:

- lotto == 0인데도 win_nums 루프를 돌고 있음 → if lotto == 0이면 continue로 건너뛰는 게 맞음 (0이 win_nums에 없으니 실제 결과엔 영향 없지만 불필요한 순회)
- 중첩 for문이 O(6×6) = 36번 — set 쓰면 더 빠름

## 풀이 패턴 정리

### 패턴 1 — `set` 교집합 활용 (가장 많은 풀이)

- `set` 교집합(`&`)으로 일치 개수를 O(n)에 계산
- `rank` 리스트를 인덱스로 쓰면 딕셔너리 불필요
- `rank[0]`, `rank[1]` 모두 6으로 설정 → 0개/1개 일치 모두 6등 처리

```python
def solution(lottos, win_nums):
    rank = [6, 6, 5, 4, 3, 2, 1]

    wins = len(set(lottos) & set(win_nums))
    zeros = lottos.count(0)

    return [rank[wins + zeros], rank[wins]]

lottos = [44, 1, 0, 0, 31, 25]
win_nums = [31, 10, 45, 1, 6, 19]
solution(lottos, win_nums)
```

### 패턴 2 — 수식으로 순위 변환

- `7 - n`으로 순위 계산, 리스트/딕셔너리 없이 수식 하나로 해결
- 단, n >= 2 조건 필요 (1개 이하는 6등이므로 `max(1, 7-n)`으로 처리 가능)

```python
def solution(lottos, win_nums):
    def get_rank(n):
        return max(1, 7 - n) if n >= 2 else 6

    zeros = lottos.count(0)
    wins = len(set(lottos) & set(win_nums))

    return [get_rank(wins + zeros), get_rank(wins)]

lottos = [44, 1, 0, 0, 31, 25]
win_nums = [31, 10, 45, 1, 6, 19]
solution(lottos, win_nums)
```
