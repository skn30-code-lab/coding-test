# 숫자 문자열과 영단어

https://school.programmers.co.kr/learn/courses/30/lessons/81301

```python
def solution(s):
    num = ['zero', 'one','two', 'three', 'four', 'five', 'six','seven', 'eight','nine']

    for i, n in enumerate(num):
        s = s.replace(n, str(i))

    answer = int(s)

    return answer


solution('one4seveneight')
```

## 내 풀이 리뷰

**장점**
- `enumerate`로 인덱스(=숫자)와 단어를 동시에 꺼내는 게 핵심 포인트. 딕셔너리 따로 안 만들어도 되는 깔끔한 구조.
- `str.replace`는 해당 단어가 없어도 그냥 넘어가므로 예외 처리 필요 없음.
- 시간복잡도 O(10n) = O(n), 실용적.

**개선 여지**
- `answer` 변수 없이 `return int(s)` 바로 써도 됨 (스타일 차이, 오답은 아님).


## 다른 풀이들

### 풀이 2 — re.sub + lambda (정규식)
정규식으로 모든 영단어를 한 번의 패턴 매칭으로 처리. `str.replace`를 10번 돌리는 대신 문자열을 1번만 순회.


```python
import re

def solution(s):
    words = ['zero','one','two','three','four','five','six','seven','eight','nine']
    pattern = '|'.join(words)  # 'zero|one|two|...|nine'
    return int(re.sub(pattern, lambda m: str(words.index(m.group())), s))

solution('one4seveneight')
```

### 풀이 3 — reduce (함수형 스타일)
`functools.reduce`로 리스트를 누산하며 문자열을 변환. 내 풀이의 for 루프를 함수형으로 표현한 것과 동일한 로직.


```python
from functools import reduce

def solution(s):
    words = ['zero','one','two','three','four','five','six','seven','eight','nine']
    return int(reduce(lambda acc, x: acc.replace(x[1], str(x[0])), enumerate(words), s))

solution('one4seveneight')
```

### 풀이 4 — 문자 단위 직접 파싱 (딕셔너리 + 버퍼)
문자열을 앞에서 한 글자씩 읽으면서 숫자면 바로 결과에 추가, 알파벳이면 버퍼에 쌓다가 딕셔너리에 매칭되는 순간 변환. 가장 로우레벨 접근이지만 replace 없이 O(n) 1회 순회로 끝남.


```python
def solution(s):
    d = {'zero':'0','one':'1','two':'2','three':'3','four':'4',
         'five':'5','six':'6','seven':'7','eight':'8','nine':'9'}
    result, buf = '', ''
    for c in s:
        if c.isdigit():
            result += c
        else:
            buf += c
            if buf in d:
                result += d[buf]
                buf = ''
    return int(result)

solution('one4seveneight')
```

## 풀이 비교 요약

| 풀이 | 핵심 기법 | 특징 |
|------|-----------|------|
| 내 풀이 | `enumerate` + `str.replace` × 10 | 가장 읽기 쉽고 직관적. 코테에서 최선 |
| 풀이 2 | `re.sub` + lambda | 문자열 1회 순회. `re` 문법 알면 간결 |
| 풀이 3 | `functools.reduce` | 내 풀이의 함수형 버전. `reduce` 연습용 |
| 풀이 4 | 딕셔너리 + 문자 버퍼 파싱 | O(n) 1회 순회, 가장 로우레벨 |

> **결론**: 코테에서는 내 풀이(풀이 1)가 정답. `re.sub`(풀이 2)는 알아두면 문자열 문제에서 강력하게 쓸 수 있음. 풀이 4는 "replace 금지" 같은 제약 조건이 붙을 때 대응 가능한 패턴.
